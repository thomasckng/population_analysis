```python
import numpy as np
from figaro.mixture import DPGMM
from figaro.plot import plot_multidim
from figaro.utils import get_priors
from tqdm import tqdm
```

```python
data = np.loadtxt('GW150914_masses.txt')
```

```python
mix = DPGMM([[10, 70], [10, 70]], prior_pars=get_priors([[10, 70], [10, 70]], data))
```

```python
n_draws = 200
draws = np.array([mix.density_from_samples(data) for _ in tqdm(range(n_draws))])
```

```python
plot_multidim(draws, samples=data, save=False, show=False)
```
