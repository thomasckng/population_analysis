```python
import bilby
import numpy as np
import matplotlib.pyplot as plt
from astropy.cosmology import FlatLambdaCDM
```

```python
z = np.linspace(0, 2, 1000)
plt.plot(z, bilby.gw.prior.UniformSourceFrame(z[0],z[-1],name='redshift', cosmology=FlatLambdaCDM(H0=0.1, Om0=0.315, Tcmb0=2.725)).prob(z), label='$H_0=0.1$')
plt.plot(z, bilby.gw.prior.UniformSourceFrame(z[0],z[-1],name='redshift', cosmology=FlatLambdaCDM(H0=1, Om0=0.315, Tcmb0=2.725)).prob(z), label='$H_0=1$')
plt.plot(z, bilby.gw.prior.UniformSourceFrame(z[0],z[-1],name='redshift', cosmology=FlatLambdaCDM(H0=10, Om0=0.315, Tcmb0=2.725)).prob(z), label='$H_0=10$')
plt.plot(z, bilby.gw.prior.UniformSourceFrame(z[0],z[-1],name='redshift', cosmology=FlatLambdaCDM(H0=30, Om0=0.315, Tcmb0=2.725)).prob(z), label='$H_0=30$')
plt.plot(z, bilby.gw.prior.UniformSourceFrame(z[0],z[-1],name='redshift', cosmology=FlatLambdaCDM(H0=70, Om0=0.315, Tcmb0=2.725)).prob(z), label='$H_0=70$')
plt.plot(z, bilby.gw.prior.UniformSourceFrame(z[0],z[-1],name='redshift', cosmology=FlatLambdaCDM(H0=100, Om0=0.315, Tcmb0=2.725)).prob(z), label='$H_0=100$')
plt.plot(z, bilby.gw.prior.UniformSourceFrame(z[0],z[-1],name='redshift', cosmology=FlatLambdaCDM(H0=130, Om0=0.315, Tcmb0=2.725)).prob(z), label='$H_0=130$')
plt.legend()
plt.xlabel('$z$')
plt.ylabel('$p(z)$')
plt.show()
```

```python
dL = np.linspace(0, 15000, 1000)
plt.plot(dL, bilby.gw.prior.UniformSourceFrame(dL[0],dL[-1],name='luminosity_distance', cosmology=FlatLambdaCDM(H0=0.1, Om0=0.315, Tcmb0=2.725)).prob(dL), label='$H_0=0.1$')
plt.plot(dL, bilby.gw.prior.UniformSourceFrame(dL[0],dL[-1],name='luminosity_distance', cosmology=FlatLambdaCDM(H0=1, Om0=0.315, Tcmb0=2.725)).prob(dL), label='$H_0=1$')
plt.plot(dL, bilby.gw.prior.UniformSourceFrame(dL[0],dL[-1],name='luminosity_distance', cosmology=FlatLambdaCDM(H0=10, Om0=0.315, Tcmb0=2.725)).prob(dL), label='$H_0=10$')
plt.plot(dL, bilby.gw.prior.UniformSourceFrame(dL[0],dL[-1],name='luminosity_distance', cosmology=FlatLambdaCDM(H0=30, Om0=0.315, Tcmb0=2.725)).prob(dL), label='$H_0=30$')
plt.plot(dL, bilby.gw.prior.UniformSourceFrame(dL[0],dL[-1],name='luminosity_distance', cosmology=FlatLambdaCDM(H0=70, Om0=0.315, Tcmb0=2.725)).prob(dL), label='$H_0=70$')
plt.plot(dL, bilby.gw.prior.UniformSourceFrame(dL[0],dL[-1],name='luminosity_distance', cosmology=FlatLambdaCDM(H0=100, Om0=0.315, Tcmb0=2.725)).prob(dL), label='$H_0=100$')
plt.plot(dL, bilby.gw.prior.UniformSourceFrame(dL[0],dL[-1],name='luminosity_distance', cosmology=FlatLambdaCDM(H0=130, Om0=0.315, Tcmb0=2.725)).prob(dL), label='$H_0=130$')
plt.legend()
plt.xlabel('$d_L$')
plt.ylabel('$p(d_L)$')
plt.show()
```
