```python
import numpy as np
import matplotlib.pyplot as plt
from figaro import plot_settings

plt.rcParams.update({
    "text.usetex": True,
    "font.family": "serif",
    "font.serif": ["Computer Modern Roman"],
})
```

```python
with np.load("./result/result_pp_plot.npz") as data:
    true_H0_arr = data['true_H0_arr']
    jsd_arr = data['jsd_arr']
    H0_samples_arr = data['H0_samples_arr']
    H0_perc_arr = data['H0_perc_arr']
```

```python
plt.figure(figsize=(6,6))
plt.plot(np.linspace(0,1,100), np.linspace(0,1,100), "k--")
plt.plot(np.sort(H0_perc_arr), np.linspace(0,1,100), "r-")
plt.title("$\mathrm{P-P\ Plot}$")
plt.show()
```

```python
for i in range(len(true_H0_arr)):
    plt.hist((H0_samples_arr[i]-true_H0_arr[i])/true_H0_arr[i], density = True, histtype = 'step', alpha = 0.2, color='C0', bins = 30)
plt.axvline(0, lw = 0.7, ls = '--', c = 'orangered', label = '$\\mathrm{Origin}$')
plt.xlabel('$(H_0^\mathrm{inferred}-H_0^\mathrm{true})/H_0^\mathrm{true}$')
plt.ylabel('$\mathrm{Density}$')
plt.xlim(-.2,.2)
plt.legend(loc = 0)
plt.show()
```
